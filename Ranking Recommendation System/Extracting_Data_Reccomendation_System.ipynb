```python
df = pd.read_csv('/content/mfp-diaries.tsv', sep='\t')
df.head()
```

```python
print(df.columns)
```

```python
df.columns = ['user_id', 'date', 'details', 'total']
```

```python
# Defining the index
i = 0

# Check if it's in the valid range
if i < len(df):
    # Define the user_id and date
    user_id = df.iloc[i, df.columns.get_loc('user_id')]
    date = df.iloc[i, df.columns.get_loc('date')]

    # Parsing JSON File
    try:
        food_entries = json.loads(df.iloc[i, df.columns.get_loc('details')])
        aggregate_intake_goal = json.loads(df.iloc[i, df.columns.get_loc('total')])
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
        food_entries, aggregate_intake_goal = [], {}

    # Print Parsing File
    print(f"User ID: {user_id}")
    print(f"Date: {date}")
    print("Food Entries:", food_entries)
    print("Aggregate Intake Goal:", aggregate_intake_goal)
else:
    print(f"Indeks {i} di luar batas DataFrame.")
```

```python
def extract_dish_sugar(user_id, date, food_entries):
    results = []
    for entry in food_entries:
        # Extracting food name
        for dish in entry.get("dishes", []):
            # Change into unknow for unknown data
            name = dish.get("name", "Unknown")

            # Calculating sum of the sugar
            sugar = next((nutri.get("value") for nutri in dish.get("nutritions", []) if nutri.get("name") == "Sugar"), 0)

            # Use the format
            results.append({
                "User ID": user_id,
                "Date": date,
                "Dish Name": name,
                "Sugar": sugar
            })

    return results
```

```python
final_results = []
for _, row in df.iterrows():  # Replace df_split with df
    user_id = row["user_id"]
    date = row["date"]
    food_entries = json.loads(row["details"])
    final_results.extend(extract_dish_sugar(user_id, date, food_entries))

# Conversion DataFrame
final_df = pd.DataFrame(final_results)
final_df['Date'] = pd.to_datetime(final_df['Date'], errors='coerce')

final_df[['Food Item', 'Serving Size']] = final_df['Dish Name'].str.split(',', expand=True, n=1)

final_df.head()
```

```python
final_df.dropna(subset=['Food Item'], inplace=True)
final_df = final_df[~final_df['Dish Name'].str.contains('Unknown')]
final_df = final_df[~final_df['Dish Name'].str.contains('Quick Added Calories')]
```
